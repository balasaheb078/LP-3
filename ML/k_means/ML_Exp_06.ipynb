```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from kneed import KneeLocator
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
```

```python
df
```

```python
numeric_data = df.select_dtypes(include=['float64', 'int64']).dropna()  # Remove non-numeric columns
```

```python
numeric_data.isnull().sum()
```

```python
X = numeric_data[['QUANTITYORDERED', 'SALES']]
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
```

```python
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (Within-cluster Sum of Squares)')
plt.title('Elbow Method to Determine Optimal k')
plt.show()
```

```python
kneedle = KneeLocator(K_range, inertia, curve="convex", direction="decreasing")
optimal_k = kneedle.elbow
print("Optimal k based on elbow:", optimal_k)
```

```python
numeric_data['Cluster'] = kmeans.labels_
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
numeric_data['Cluster']=kmeans.fit_predict(X)
```

```python
centroids = kmeans.cluster_centers_
```

```python
print(numeric_data[['QUANTITYORDERED', 'SALES', 'Cluster']])
```

```python
plt.scatter(numeric_data['SALES'], numeric_data['QUANTITYORDERED'], c=numeric_data['Cluster'])  # Swapped x and y
plt.scatter(centroids[:, 1], centroids[:, 0], marker='.', s=200, linewidths=3, color='r')  # Swapped centroid coordinates
plt.xlabel('SALES')
plt.ylabel('QUANTITYORDERED')
plt.title('K-Means Clustering with Centroids')
plt.show()
```
